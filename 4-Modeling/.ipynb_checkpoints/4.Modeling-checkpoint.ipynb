Given an overall about the metrics:

1. Accuracy: If the predicted label match the true label, the acuracy is 1 for the sample. The sum of correct predictions over n samples is the accuracy. 
2. Precision: Is the ability of the classifier not to label as positive a sample that is negative.The formulation is tp / (tp + fp) where tp is the number of true positives and fp the number of false positives
3. Recall: Is the ability of the classifier to find all the positive samples. he recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives.
4. F1: Harmonic meaan of the precision and recall. F1 = 2 * (precision * recall) / (precision + recall).

### D) Feature Selecting

We created some interesting features, now we will select de best set of features. We have 15 features, is impracticable test all the features, luckly there is a number of procedures to choose. The logistic regression has reliable coeficients for the features, we can use the coeficient to eliminate recursively the best set. The algorithm name is Recursive Feature Elimination.

The procedure Recursive feature elimination(RFE) starts with all the features available and will recursively eliminating the features with lowest impact based on the coeficients. The procedure will repeat util the best set is determined. 

```python
from sklearn.feature_selection import RFECV

lr_model = LogisticRegression(**lr_parameters)
selector = RFECV(lr_model, step=1, cv=10, scoring = scorer[0], n_jobs=-1)
selector = selector.fit(x_train, y_train['Survived'].to_numpy())

selector.ranking_
```

We can try another models, let's see the performance for an ensemble model, a variant of the random forest, the extremely randomized trees(Extra-trees). The extra-trees modify the random forest to make then more random, what justify the name extremely randomized. The default extra-trees doesn't limit the grow of the trees instead the method choose to increase the variance between the trees by increasing the randomness on the procedure of growth. More especificaly, the cutting points in the nodes is choosed randomly and evaluated by a metric, the best cutting point is selected. The default setup on sklearn doesn't use bootstrap(typical procedure on forests, which replace the original data-set with a replica of the data set), but we can choose to use. 

### E)Other Method (Extra-Trees)

```python
from sklearn.ensemble import ExtraTreesClassifier
#Loading the original dataset
x_train_orig = pd.read_csv(f"{data_path}/x_train.csv", index_col = 0)
x_test_orig = pd.read_csv(f"{data_path}/x_test.csv", index_col = 0)

#Selecting the x_train_best
best_features = x_train.columns[selector.support_]
x_train_best = x_train[best_features]

#Modeling the best set.
et_model_best = ExtraTreesClassifier(n_estimators=2000, bootstrap=True , n_jobs=-1, min_samples_split=3 , min_impurity_decrease= 0.00001)
cr = cross_validate(et_model_best, x_train_best, y_train['Survived'].to_numpy(), cv=10, scoring = scorer)
cr_values_best = list(cr.values())[2:]
cr_keys_best = [f"{key}_best" for key in list(cr.keys())[2:]]

#Modeling the original.
et_model_orig = ExtraTreesClassifier(n_estimators=2000, bootstrap=True, n_jobs=-1,  min_samples_split=3 ,min_impurity_decrease= 0.00001)
cr = cross_validate(et_model_orig, x_train_orig, y_train['Survived'].to_numpy(), cv=10, scoring = scorer)
cr_values_orig = list(cr.values())[2:]
cr_keys_orig = [f"{key}_orig" for key in list(cr.keys())[2:]]

cr_values =  [[cr_best, cr_orig] for cr_best, cr_orig in zip(cr_values_best, cr_values_orig)]
cr_keys = [[cr_best, cr_orig] for cr_best, cr_orig in zip(cr_keys_best, cr_keys_orig)]
cr_val_list = []
for cval in cr_values:
    for cv in cval:
        cr_val_list.append(cv.ravel())
        
cr_keys = np.asarray(cr_keys).ravel()
fig1, ax1 = plt.subplots(figsize=(16, 8))
ax1.set_title('Baseline Boxplot')
bplot = ax1.boxplot(cr_val_list ,labels = cr_keys, patch_artist = True, showmeans =True)
ax1.yaxis.grid(True)
# fill with colors
colors = ['pink','deeppink', 'lightblue', 'deepskyblue', 'lightgreen', 'darkseagreen', 'bisque', 'sandybrown']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.plot()
```

The results is slight better compared to the logistic regression. This method also has a feature importance which we can use. The feature importance of the logistic regression is the coeficients on the equation, it's a pretty straightfoward way to acquire feature importance. On contrast, the feature importance on the extra-trees is a calculation on the nodes of the trees. The tree methods use a tree structure to classify the samples, each node of the tree has cutting on a feature, i.e., feature_1 <5, if true move the sample to right, else move the sample to left. The eficiency of the cut can be expressed by how much improvement is made by the cut rule. A cutting rule which separate perfectly the samples is better than one which separete the samples partially. This can generate a score and summing this score for each node we have the feature importance. So let's procced.

```python
#modeling
et_model = ExtraTreesClassifier(n_estimators=2000, bootstrap=True , n_jobs=-1, min_samples_split=3 , min_impurity_decrease= 0.00001)
et_model = et_model.fit(x_train, y_train['Survived'].to_numpy())
et_model.feature_importances_
et_model.feature_names_in_
rank_features = {name: rank for name, rank in zip(et_model.feature_names_in_, et_model.feature_importances_)}
#sorting
rank_features = dict(sorted(rank_features.items(), key=lambda item: item[1], reverse=True))

fig1, ax1 = plt.subplots(figsize = (18, 6))
ax1.set_title('Features rank')
bar = ax1.bar(rank_features.keys(),rank_features.values())
plt.xticks(rotation=80)
plt.tight_layout()
plt.plot()
```

The next method is one of my favorites. The procedure name is Sequential Feature Selector (SFS). It's a simple test where we can beggin in two diferent ways: We can start with one feature and test all possibles 1 feature sets, or we can start by setting the max possible set of features. The former is called forward way, and at each iteration one new feature is added on the group. In contrast, the second way is the backward, at each iteration one feature is removed. The algorithm stops when there is no improvement. The advantage on this procedure is that the features may interfer on others. On the other hand, the algorithm may test a great number of features and is not so fast than the other procedures. Besides the algorithm does not assure encouter the best feature set. As we already said, the features interfer in one another, maybe the best set is revealed after adding two features ahead, but adding one does not give the best set in the interation, so the best set will not be found.

```python
from sklearn.feature_selection import SequentialFeatureSelector
et_model = ExtraTreesClassifier(n_estimators=500, bootstrap=True, n_jobs=-1,  min_samples_split=3 ,min_impurity_decrease= 0.00001)


#Modeling the best set.
sfs_lr_fwrd = SequentialFeatureSelector(lr_model, cv=split, scoring = 'f1', n_jobs=-1, direction='forward').fit(x_train, y_train['Survived'].to_numpy())
print('End sfs lr fwrd!')
sfs_lr_bkrd = SequentialFeatureSelector(lr_model, cv=split, scoring = 'f1', n_jobs=-1, direction='backward').fit( x_train, y_train['Survived'].to_numpy())
print('End sfs lr bkrd!')

sfs_et_fwrd = SequentialFeatureSelector(et_model, cv=split, scoring = 'f1', n_jobs=-1, direction='forward').fit(x_train, y_train['Survived'].to_numpy())
print('End sfs et fwrd!')
sfs_et_bkrd = SequentialFeatureSelector(et_model, cv=split, scoring = 'f1', n_jobs=-1, direction='backward').fit( x_train, y_train['Survived'].to_numpy())
print('End sfs et bkrd!')
```

### H)Enseble the models

Now that we have three classifiers, we can try to combine the prediction of each of them. The idea is that we have different methods that maybe is good for some samples but for others not. We can try to sum the best part of each model and make a final strong classifier.

First, we will start with a simple approach, just stacking the classifiers logistic regression, extra-trees and SVM and choosing the class by majority vote. Them we will try to combine the result in a better fashion. Note that this approach is proned to overfitting, trying to avoid this we will also use cross validation techiniques.

```python
from sklearn.ensemble import VotingClassifier
#instanciating the models
lr_model = LogisticRegression(**lr_parameters)
svc_model = SVC(C=c, probability=True, class_weight={0:0.3, 1:1})
et_model = ExtraTreesClassifier(n_estimators=2000, bootstrap=True , n_jobs=-1, min_samples_split=3 , min_impurity_decrease= 0.00001, class_weight={0:1, 1:0.5})

#ensembling the models
```

```python

#x_train_best_etfeat = x_train[best_features_et]
#x_train_best_lrfeat = x_train[best_features_lr]

w = [1,1,1]
#Modeling the best set extra trees features
eclf = VotingClassifier(estimators=[('lr', lr_model), ('extra-trees', svc_model), ('SVM-RBF', et_model)],voting='soft', n_jobs=-1, weights= w)
cr = cross_validate(eclf, x_train_best_etfeat, y_train['Survived'].to_numpy(), cv=10, scoring = scorer)
cr_values_best_etfeat = list(cr.values())[2:]
cr_keys_best_etfeat = [f"{key}_best_etfeat" for key in list(cr.keys())[2:]]

#Modeling the best set extra trees features
eclf = VotingClassifier(estimators=[('lr', lr_model), ('extra-trees', svc_model), ('SVM-RBF', et_model)],voting='soft', n_jobs=-1, weights= w)
cr = cross_validate(eclf, x_train_best_lrfeat, y_train['Survived'].to_numpy(), cv=10, scoring = scorer)
cr_values_best_lrfeat = list(cr.values())[2:]
cr_keys_best_lrfeat = [f"{key}_best_lrfeat" for key in list(cr.keys())[2:]]


#Modeling the original.
eclf = VotingClassifier(estimators=[('lr', lr_model), ('extra-trees', svc_model), ('SVM-RBF', et_model)],voting='soft', n_jobs=-1, weights= w)
cr = cross_validate(eclf, x_train_orig, y_train['Survived'].to_numpy(), cv=10, scoring = scorer)
cr_values_orig = list(cr.values())[2:]
cr_keys_orig = [f"{key}_orig" for key in list(cr.keys())[2:]]

cr_values =  [[cr_best_lrfeat, cr_best_etfeat, cr_orig] for cr_best_lrfeat, cr_best_etfeat, cr_orig in zip(cr_values_best_lrfeat, cr_values_best_etfeat, cr_values_orig)]
cr_keys = [[cr_best_lrfeat, cr_best_etfeat, cr_orig] for cr_best_lrfeat, cr_best_etfeat, cr_orig in zip(cr_keys_best_lrfeat, cr_keys_best_etfeat, cr_keys_orig)]
cr_val_list = []
for cval in cr_values:
    for cv in cval:
        cr_val_list.append(cv.ravel())

cr_keys = np.asarray(cr_keys).ravel()
fig1, ax1 = plt.subplots(figsize=(16, 10))
ax1.set_title(f'Boxplot Ensemble of Classifiers best sets: lr, et, origi')
bplot = ax1.boxplot(cr_val_list ,labels = cr_keys, patch_artist = True, showmeans =True)
ax1.yaxis.grid(True)
plt.xticks(rotation=15)
# fill with colors
colors = ['pink','deeppink','hotpink',  'lightblue', 'deepskyblue', 'royalblue', 'lightgreen', 'olivedrab','darkolivegreen', 'bisque', 'sandybrown', 'sienna']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.plot()
```

This simple method doesn't perform better than the most of classifiers alone. Maybe the reason is that the classifiers errors is pretty much the same, and when the classifiers miss the class the other one also miss. Trying to correct this, we will use the stacking procedure. The stacking puts another layer on the classifiers. The new layers, sometimes called meta, trains another machine learning algorithm with the stacking responses. The new layer resposability is decide when to use each of the models stacked. 

```python
from sklearn.ensemble import StackingClassifier,AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB
estimators = [('lr', lr_model), ('SVM-RBF', svc_model), ('et_model', et_model), ('AdaBoost',AdaBoostClassifier(n_estimators=2000)), ('NaiveBayes-Gaussian',GaussianNB()), ('Bernoulli-nb',BernoulliNB())]
stack_clf = StackingClassifier(estimators=estimators, final_estimator=ExtraTreesClassifier(n_estimators=2000, bootstrap=True , n_jobs=-1), n_jobs=-1, passthrough=True)

x_train_best_etfeat = x_train[best_features_et]
x_train_best_lrfeat = x_train[best_features_lr]

#Modeling the best set extra trees features
cr = cross_validate(stack_clf, x_train_best_etfeat, y_train['Survived'].to_numpy(), cv=10, scoring = scorer, n_jobs=-1)
cr_values_best_etfeat = list(cr.values())[2:]
cr_keys_best_etfeat = [f"{key}_best_etfeat" for key in list(cr.keys())[2:]]

#Modeling the best set extra trees features
cr = cross_validate(stack_clf, x_train_best_lrfeat, y_train['Survived'].to_numpy(), cv=10, scoring = scorer, n_jobs=-1)
cr_values_best_lrfeat = list(cr.values())[2:]
cr_keys_best_lrfeat = [f"{key}_best_lrfeat" for key in list(cr.keys())[2:]]


#Modeling the original.
cr = cross_validate(stack_clf, x_train_orig, y_train['Survived'].to_numpy(), cv=10, scoring = scorer, n_jobs=-1)
cr_values_orig = list(cr.values())[2:]
cr_keys_orig = [f"{key}_orig" for key in list(cr.keys())[2:]]

cr_values =  [[cr_best_lrfeat, cr_best_etfeat, cr_orig] for cr_best_lrfeat, cr_best_etfeat, cr_orig in zip(cr_values_best_lrfeat, cr_values_best_etfeat, cr_values_orig)]
cr_keys = [[cr_best_lrfeat, cr_best_etfeat, cr_orig] for cr_best_lrfeat, cr_best_etfeat, cr_orig in zip(cr_keys_best_lrfeat, cr_keys_best_etfeat, cr_keys_orig)]
cr_val_list = []
for cval in cr_values:
    for cv in cval:
        cr_val_list.append(cv.ravel())

cr_keys = np.asarray(cr_keys).ravel()
fig1, ax1 = plt.subplots(figsize=(16, 10))
ax1.set_title(f'Boxplot stacking of Classifiers best sets: (lr, et, origi)lr')
bplot = ax1.boxplot(cr_val_list ,labels = cr_keys, patch_artist = True, showmeans =True)
ax1.yaxis.grid(True)
plt.xticks(rotation=15)
# fill with colors
colors = ['pink','deeppink','hotpink',  'lightblue', 'deepskyblue', 'royalblue', 'lightgreen', 'olivedrab','darkolivegreen', 'bisque', 'sandybrown', 'sienna']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
plt.plot()
```
